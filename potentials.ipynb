```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def V(x, s, d):
    return (x**2-s**2)**2 + s*d*(1.0/3*x**3 - s**2*x)
# Potential parameters
s = 2.2
d = 0.1
# Discretisation
L = 3.4
nx = 100
xedges = np.linspace(-L, L, nx+1)
dx = xedges[1]-xedges[0]
xcenters = xedges[0:-1]+0.5*dx
plt.plot(xcenters, V(xcenters, s, d))
```

```python
%matplotlib inline

import numpy as np
import scipy as sp
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
# Import the data
df_train = pd.read_csv('kaggle_house_prices/train.csv')
```

```python
df_train.columns
```

```python
df_train['SalePrice'].describe()
```

```python
plt.figure(figsize=(10,5))
sns.set_context("poster")
sns.set_style('darkgrid')
sns.distplot(df_train['SalePrice'], kde=False)
plt.ylabel('Count')
plt.xlabel('Sale Price ($)')
```

```python
plt.figure(figsize=(8,7))
sns.set_context('poster')
plt.scatter(df_train['GrLivArea'], df_train['SalePrice'])
```

```python
from sklearn.linear_model import LinearRegression

regr = LinearRegression()
x = np.array([df_train.GrLivArea.get_values()])
y = np.array([df_train.SalePrice.get_values()])

regr.fit(x.transpose(),y.transpose())
plt.scatter(x[0],y[0], c='green')
plt.plot(x[0], regr.predict(x.transpose()), lw=4, c='Red')
```
